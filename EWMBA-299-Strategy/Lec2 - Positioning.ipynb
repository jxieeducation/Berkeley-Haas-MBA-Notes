## Walmart case
* Walmart's business seems very simple to immitate. It's just a retail store.
* There are winners (e.g. Target, Kmart), but there are a long tail of losers in this space.
* why did Walmart win?
    * horizontal positioning (going to rural areas)
    * operated w/ lower costs
        * mom and pop bought from wholesalers, but Walmart bought directly from manufacturers
        * Walmart employed a Hub-and-Spoke distribution system
        * higher capacity utilization -> point of sale data capture
            * trucks on the way back to warehouse would pick up inventor from suppliers
                * **like Uber!! more capacity utilization **
        * higher turn over (shelf)
        * no storage space in store (more display, less shrinking from employee theft)
        * only buy branded products to save marketing
    * **walmart was a tech company of its day!!!**

## Positioning and Competitive Advantage

### External Approach
* Industry analysis - strategy management concerned with
    * locating attractive industries and segments
    * position your firm within that industry
    * work to modify industry structure
    * work to modify competitor behavior

### Ricardian Rents
* access to exclusive resources (e.g. soil great for growing wine!)
* **can't copy everything and compete blindly**

### Internal Approach
* resource view emphasizes uniqueness of firm
* key to profitability is not doing what others do (benchmarking) but exploiting differences
* firms that have unique combinations of resources and capaibilities are difficult to imitate

## Strategy Formulation

### External Change
* one approach: definite markets broadly
* examples:
    * UPS in logistics
    * Xerox in document processing
    * Railroad companies in transportation (e.g. flights)
* firms capabilities might not transfer well to new products

### Internal Change
* Rather than following customer needs, focus on what the firm does well
* Example
    * Railroad companesi into pipelines, telecoms and logistics
    * 3M adhesives and thin-hilm coating

### examples
* Honda - diesel to electricity generator -> bike -> car -> luxury car
* Remmington - typewriters -> razors (assembling compelx things)
* Microplane - wood sculpting tools -> vegetable slicing tool
* Google - search -> self-driving cars (ML)

### Gillette vs BiC
* Gillette had good disposable razors
* BiC (pens bisness) went to compete
* Gillette competed with BiC by attacking their core business (pens - PaperMate)
* what they could have done
    * buy out - no, can't do it for everyone
    * compete price - losing proposition
    * attack their core business instead

* shifting the focus of strategy analysis
    * from external to internal environment
