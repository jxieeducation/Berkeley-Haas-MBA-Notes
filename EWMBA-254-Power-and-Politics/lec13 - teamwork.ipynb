### Social motivation
* social facilitation - audience effect
    * other ppl make you perform better, by having an audience / accountability
* social loafing 
    * free riding
        * no neg consequences for loafing
        * postive consequences occur regardless
    * sucker effect
        * don't want to be taken adv
* (choking)

### Stop social loafing
1. identifiable inputs (cant get away w/ it because others will know)
    * individual performance measured
    * smaller team size
    * proximity
    * specialization
2. indispensibility
    * how much does your input matter?
    * cant get away with it bc it will occur
    * need to make people know that +/- consequences do occur depending on individual effort
3. trust
    * calculative trust (extrinsic)
        * anything that reduces free riding
    * affective trust
        * anything that allows members to know and trust each other
        
(1 and 2 solves social loafing, 3 solves sucker effect)

## Stages of group dev 

### 5 stage model
* forming (if you want to be the leader, this is the time to start)
* storming
* norming 
* performing
* adjourning

### Punctuated Equilibrium Model
* start
* mid point transition (if you missed the chance to grab leadership in the start, now is the last chance!)
    * e.g. oh shoot, assignment is due next week
* end

### Heuristic model of performance
* performance = energy * focus * opportunity + luck_controllable + luck_random
* energy = motivation, law of effect, intrinsic / extrinsic
* focus = how motivation and other resources are used
* opportunity includes resources
* luck_controllable means being in the right place at the right time, knowing the right people...etc.

### Cohesion
* group size
* required interaction patterns (freq of interactions)
* success
* homogeneity (ppl like others who remind them on themselves)
* competition / external threat
* reward system
* difficulty of entrance (e.g. selectiveness, hazing)

### Effects of Cohesiveness
* highly cohesive grps enforce norms
* punish deviants
* memebers dont want to be punished / kicked out
* more cohesiveness means that members increasingly behaves the same
