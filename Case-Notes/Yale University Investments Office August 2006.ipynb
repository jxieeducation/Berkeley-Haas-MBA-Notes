## David Swensen and the Investment office in 2006
* Swensen was a Yale economics phd, went to Lehman, but came back because he also wanted to teach
* 20 professionals in 2006
* situated close to Yale
* reported to president and endowment office
* also oversaw spends on endowment

## Investment Philosophy
1. Believed strongly in equity
    * equities have a claim on real stream of income
    * bonds perform poorly when the market is bull or inflation is high
2. Diversified portfolio
    * avoid timing the market
    * even if they had their own views, they think most of the time, its already reflected in the market
3. Seek opportunities in less efficient markets
    * The difference between the 75th and 25th quantile fund managers differ more and more widely when the market is more inefficient
4. Use outside fund managers
    * external advisors would have autonomy to implement their strategy
    * Investment office staff needed to develope close and mutually beneficial relationships
5. Align manager interests with the office's
