## 2. Blockchain: From the practice of the incumbent to that of the informal

### 2.1 Cash Preferences 
* cash transactions are customary among financially excluded individuals
    * e.g. salary, agriculture products
* in-cash savings mechanisms are also preferred in the form of peer-to-peer local transaction
* for unbanked individuals, cash is a competition to blockchain
* mobile banking - blockchain can piggyback off of that

### 2.2 Lending Practices
* rotating savings and credit association (ROSCAs) - participants commit to a common pot, each member receives full pot on one occasion
* blockchain can lower transactional costs when individuals in informal networks are exploited by local usury or led into bonded labor
    * WeTrust - trusted lending circle
        * "WeTrust ROSCA allows users to save and issue credit as a group, at self-determined and fair interest rates. "
        * more here: https://www.facebook.com/pg/wetrustplatform/about/?ref=page_internal
    * MoneyCircles
        * borrow from people you know

### 2.3 Transfers and Remittances
* prepaid debit cards 
* mobile banking also good for transfers
* avoids the need for a bank acc is great!
    * Coins.ph - pay bills...etc.
    * bluezelle.com - decentralized storage of data
    * can also work for insurance - street vendors collectively utilize a mutual insurance pool through a distributed ledger
* fast for remittances!
* sensitive to local networks

### 2.4 Identification Services
* reputation based peer to peer lending 
    * previous behavior informs future
* proof of existence
