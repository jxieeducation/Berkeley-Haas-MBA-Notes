* startups need more than product development and marketing, they also need customer discovery and validation
* marketers have developed tools, typically to introduce new products
* entrepreneurs need more market research in the traditional sense
    * a. coherent and falsible hypotheses about relevant variables
    * b. timely and affordable means for investigating the hypotheses

## What do you want to know about potential customers?
* a hypothesis is more than an assumption, it's a testible proposition
* startups due to resource constraint need to be very disciplined about:
    * customer problem and product
        * initial hypothesis about a customer's problem and venture's solution
        * core to the value proposition in the business model
    * demand creation and buying process
        * initial hypothesis about how potential customers will find the product
        * core to the go-to-market component

### Customer problem and product
* verify that a market could exist for the proposed offering
* determine whether a product should be built
#### problems vs products
* no one pays for a product, they pay for a satisfactory solution to their problem
* you are not a product, but the problem that you solve
* e.g. railroad (product) vs transportation (problem)
* Steve Blank emphasises a customer development process, to run in parallel with the product development process
* verify customer demand for a proposed solution
* during customer development, it is important to understand the value proposition, confirming that there's an unmet need

#### benefits vs status quo
* people perceive loss / cost more than the benefit 
* understand how customers perceive costs like: transaction cost, learning cost, obsolescence cost

#### lead user vs mainstream user
* visionaries vs pragmatists
* lead users have these 4 properties
    * has a problem
    * understands that it's a problem
    * actively searching for a soln
    * has an interim, home-grown soln
* make hypothesis about who see your soln as providing far superior value than mainstream

### Demand creation and buying process
* rows of people in enterprise:
    * initiator - recognize problem
    * gatekeeper - control information
    * influencer - those who have a say in the purchase
    * decider
    * purchaser
    * users


## How can you validate demand?
* quatative questions - need to be asking the right question
* qualitative questions - probing, understanding problems
* qualitative research surfaces questions answered by quantative research
* anomalies in quantative research are explored by qualitative research
* entrepreneurs should not rely too much on published data, markets are rarely precisely defined
* investors prefer to see real data on customer relations

### Customer surveys
* better suited for invalidating hypothesis then validating
* cons
    * bundled answers
    * do not expose loss aversion 
* pros
    * cheap
### letter of intent
* customer agrees to try / buy a product
* non legally binding, but still good to guage
* if few prospects will sign LOI, then customer demand has been disconfirmed
* can learn about customer needs by negotiating LOI terms

### usability test
* goal is to rule out the risk that a flawed MVP design will lead an entrepreneur to conclude weak demands
* if early demand is weak, it's important to know:
    1. is there no need for the problem?
    2. want a soln, but can't figure out how to use it
* small number of users is good enough to discover most flaws
* e.g. try usertesting.com

### market trials
* launch with a small portion e.g. 1 city
* can figure out if demand is there by charging full price
* see if a customer's LTV is positive, net the CAC

### split tests
* ab testing
* for early companies, usually big changes e.g. pricing

### net promoter score
* ask customers how likely they are to promote it to a friend
* NPS is the percent of promoters minus percent detractors
* can track NPS to avoid premature scaling
* NPS varies by industry, 60 to 80 is high
* can also focus on relative NPS, need to be 10 or 20 pts higher than competitor
